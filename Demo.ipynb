```python
import os
import glob
from PIL import Image
from transformers import LlavaNextProcessor, LlavaNextForConditionalGeneration
import torach

def generatePrompt(imagenet_path=r'.\datasets\imageNet_images\imageNet_images'):

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    processor = LlavaNextProcessor.from_pretrained("llava-hf/llava-v1.6-mistral-7b-hf")
    model = LlavaNextForConditionalGeneration.from_pretrained(
        "llava-hf/llava-v1.6-mistral-7b-hf",
        torch_dtype=torch.float16,
        low_cpu_mem_usage=True
    )
    model.to(device)
    
    classMap = {'n02106662': 'German shepherd', 'n02124075': 'Egyptian cat', 'n02281787': 'Lycaenid butterfly',
                        'n02389026': 'Sorrel horse', 'n02492035': 'Capuchin', 'n02504458': 'Elephant',
                        'n02510455': 'Panda', 'n02607072': 'Anemone fish', 'n02690373': 'Airliner', 
                         'n02906734': 'Broom', 'n02951358': 'Canoe', 'n02992529': 'Cellphone',
                        'n03063599': 'Mug', 'n03100240': 'Convertible', 'n03180011': 'Desktop PC', 
                        'n03197337': 'Digital watch', 'n03272010': 'Electric guitar', 'n03272562': 'Electric locomotive',
                        'n03297495': 'Espresso maker', 'n03376595': 'Folding chair', 'n03445777': 'Golf ball', 
                        'n03452741': 'Piano', 'n03584829': 'Iron', 'n03590841': 'Jack-oâ€™-lantern', 
                        'n03709823': 'Mailbag', 'n03773504': 'Missile', 'n03775071': 'Mitten', 
                        'n03792782': 'Mountain bike', 'n03792972': 'Mountain tent', 'n03877472': 'Pyjama', 
                        'n03888257': 'Parachute', 'n03982430': 'Pool table', 'n04044716': 'Radio telescope', 
                        'n04069434': 'Reflex camera', 'n04086273': 'Revolver', 'n04120489': 'Running shoe', 
                         'n07753592': 'Banana', 'n07873807': 'Pizza', 'n11939491': 'Daisy', 'n13054560': 'Bolete'
                        }
    dir_list1 = os.listdir(imagenet_path)
    dir_list = sorted(dir_list1)
    # print(dir_list)
    c=0
    for fold in dir_list:
        image_names = os.listdir(os.path.join(imagenet_path,fold))
        for image_name in image_names:
            image_path = os.path.join(imagenet_path, image_name.split('_')[0], image_name)
            class_fname = image_name.split('_')[0]
            if class_fname in classMap.keys():
                class_name = classMap[class_fname]
            else:
                raise ValueError(f'\'{class_fname}\' is not present as a key in the class mapping dict')
        
            image_original = Image.open(image_path).convert('RGB') 

            image = Image.open(image_path).convert('RGB')
            prompt = f"[INST] <image>\nDescribe this image  and the {class_name} altogether in one line? [/INST]"
            
            inputs = processor(prompt, image, return_tensors="pt").to(device)
            output = model.generate(**inputs, max_new_tokens=100)

            img_desc = processor.decode(output[0], skip_special_tokens=True)
            with open(r'.\datasets\imageNet_images\imageNet_descriptions.txt', "a") as file:
                txt = image_name + " : " + img_desc +"\n"
                file.write(txt)
            file.close()
        c+=1
        print("Class "+ str(c) + " done.")
```
